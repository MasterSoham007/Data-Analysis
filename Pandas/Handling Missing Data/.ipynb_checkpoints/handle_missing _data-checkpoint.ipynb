```python
# convert day column from string to date type
df = pd.read_csv("weather_data.csv", parse_dates = ["day"])
df
```

```python
type(df.day[0])
```

```python
# set day column as index
df.set_index('day', inplace = True)
df
```

```python
# fill NaN values with 0
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
# if NaN value is encountered fill forward with previous value
new_df = df.fillna(method = 'ffill')
new_df
```

```python
# backward fill method
new_df = df.fillna(method = 'bfill')
new_df
```

```python
# backward value w.r.t. columns
new_df = df.fillna(method = 'bfill', axis = "columns")
new_df
```

```python
# if NaN value is encountered fill forward all corresponding cells with previous value
new_df = df.fillna(method = 'ffill')
new_df
```

```python
# forward fill only once
new_df = df.fillna(method = 'ffill', limit = 1)
new_df
```

```python
df
```

```python
# interpolate method user linear interpolation to come up with better values
# The linear interpolation formula is the simplest method that is used for estimating the value of a function between any two known values.
new_df = df.interpolate()
new_df
```

```python
# interpolates keeping the time parameter in mind
new_df = df.interpolate(method = 'time')
new_df
```

```python
# drop all rows with n.a. values

new_df = df.dropna()
new_df
```

```python
# drop rows where all cells have n.a. values

new_df = df.dropna(how = 'all')
new_df
```

```python
# if we have atleast one non n.a. value keep the row

new_df = df.dropna(thresh = 1)
new_df
```

```python
# if we have atleast two non n.a. value keep the row

new_df = df.dropna(thresh = 2)
new_df
```

```python
# insert missing dates
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data2.csv")
df
```

```python
new_df = df.replace([-99999, -88888] , np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : [-99999, -88888],
    'event' : '0'
},np.NaN)
new_df
```
