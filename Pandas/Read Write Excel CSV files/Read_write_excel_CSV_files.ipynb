```python
# eps can be negative but revenue cannot be negative.....if we use the above mentioned method the eps will alsp
# become NaN.To solve this we can use dictionary as input to na_values.

df =pd.read_csv("stock_data.csv",na_values = {
    "eps" : ["not available" , "n.a."],
    "revenue" : ["not available" , "n.a." , -1],
    "people" : ["not available" , "n.a."],
    "price" : ["not available" , "n.a."]
})
df
```

Excel Files

```python
import pandas as pd

df = pd.read_excel("stock_data.xlsx", "Sheet1")
df
```

```python
# convert n.a. to sam walton

def convert_people_cell(cell):
    if cell == "n.a.":
        return "sam walton"
    return cell

def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell


df = pd.read_excel("stock_data.xlsx" , "Sheet1" , converters = {
    'people' : convert_people_cell,
    'eps' : convert_eps_cell
})

df
```

```python
# write the above dataframe to a new excel file
df.to_excel("stock_data2.xlsx" , "stocks")
```

```python
# write the above dataframe to a new excel file starting from 2 nd row and 3 rd column.
# indexing from 0.
df.to_excel("stock_data3.xlsx" , "stocks", startrow = 1, startcol = 2)
```

```python
# write the above dataframe to a new excel file
df.to_excel("stock_data4.xlsx" , "stocks" , index = None , header = None)
```

```python
# writing two dataframes into a single excel file
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})

with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
